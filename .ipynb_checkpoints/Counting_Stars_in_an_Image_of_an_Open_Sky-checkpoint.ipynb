```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from math import sqrt
from skimage.color import rgb2gray
import glob
from skimage.io import imread
```

```python
#example_file = glob.glob(r"./wint_sky.gif")[0] #example1

#example_file = glob.glob(r"./200_d.gif")[0]   #example2

example_file = glob.glob(r"./MaleFeistyAmericancrocodile-size_restricted.gif")[0] #example3
```

```python
im = imread(example_file, as_grey=True)
plt.imshow(im, cmap=plt.cm.gray)
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=.1)
```

```python
#Compute radii in third column.
blobs_log[:,2] = blobs_log[:,2] *sqrt(2)
numrows = len(blobs_log)
print("Number of stars counted : ",numrows)
```

```python
fig, ax = plt.subplots(1,1)

plt.imshow(im, cmap=plt.cm.gray)
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+5, color='lime',linewidth=2,fill=False)
    ax.add_patch(c)
```
